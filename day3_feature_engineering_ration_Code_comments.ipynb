# Age(나이차)

```python
#결측치를 -99로 채웠기 때문에 그냥 계산하면 -인 값들이 나온다.
#이에 -99가 결측치를 채워준 부분이 등장시 나이차도 -99로 표시하게끔 첫번째로 정의한다.

def age_func(x):
    if x['age'] == -99:
        return -99
    elif x['age_o'] == -99:
        return -99
    elif x['gender'] == 'female':
        return x['age_o'] - x['age']
    else:
        return x['age'] - x['age_o']
```

```python
dating_df['age_gap'] = dating_df.apply(age_func, axis = 1)
```

```python
dating_df['age_gap']
```

```python
dating_df['age_gap_dir'] = dating_df['age_gap'].apply(lambda x: 'positive' if x > 0 else 'negative' if x < 0 else 'zero')
```

```python
dating_df['age_gap'] = abs(dating_df['age_gap'])
```

```python
def rating(data, important, score):
    if data[score] == -99:
        return -99
    elif data[important] == -99: #영상에서 추후 해당 코드를 나중에 추가 합니다.
        return -99 #영상에서 추후 해당 코드를 나중에 추가 합니다.
    else:
        return data[important] * data[score]
```
