### 결제 시간 feature 분석

```python
cc_df.info()
```

```python
cc_df['trans_date_trans_time'] = pd.to_datetime(cc_df['trans_date_trans_time'])
```

```python
from datetime import datetime 
```

```python
cc_df['hour'] = cc_df['trans_date_trans_time'].dt.hour
```

```python
cc_df.head()
```

```python
def hour_func(x):
    if (x >= 6) & (x < 12): ## 괄호! 연산자! 
        return 'morning'
    elif (x >= 12) & (x < 18): 
        return 'afternoon'
    elif (x >= 18) & (x < 23): 
        return 'evening'
    else:
        return 'night'
```

```python
cc_df['hour_cat'] = cc_df['hour'].apply(hour_func)
```

```python
cc_df.head()
```
